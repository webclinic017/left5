```python
import scrapy
from scrapy.crawler import CrawlerProcess
```

```python
class AmazonSpider(scrapy.Spider):
    name = 'amazonspider'
    start_urls = ['https://www.amazon.com/s?k=Wallets']
    custom_settings = { 
      'https': 'scrapy.core.downloader.handlers.http2.H2DownloadHandler',
   }
    
    def parse(self, response):
        for asin in response.xpath('//div[contains(@class, "s-result-list")]/div[contains(@data-component-type, "s-search-result")]'):
            yield {'asin': asin.xpath('@data-asin').extract()}

        for next_page in response.xpath('//div[contains(@class, "s-pagination-container")]//a[contains(@aria-label, "next page")]'):
            yield response.follow(next_page, self.parse)
```

```python
process = CrawlerProcess({
    'USER_AGENT': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/102.0.0.0 Safari/537.36',
    'https': 'scrapy.core.downloader.handlers.http2.H2DownloadHandler',
})
```

```python
process.crawl(AmazonSpider)
process.start()
```
