```python
import os
import urllib
import urllib2
import urlparse
from lutils.lrequest import LRequest
```

```python
lr = LRequest()
```

```python
# lr.load('http://faculty.chicagobooth.edu/ruey.tsay/teaching/fts3/')
lr.load_file('ruey_tsay.html')
```

```python
p_eles = lr.xpaths('//body/p')
```

```python
data_root = 'data'

chapter = None
for p_ele in p_eles:
    ele = p_ele.xpath('.//*[contains(text(), "Chapter")]')
    
    if len(ele) > 0:
        chapter = ele[0].text.split(':')[0]
    
    if chapter:
        _path = os.path.join(data_root, chapter)
        if not os.path.exists(_path): os.makedirs(_path)
        
        a_eles = p_ele.xpath('.//a')
        for a_ele in a_eles:
            
            name = a_ele.text if a_ele.text and a_ele.text.strip() else a_ele.attrib['href'].rsplit('/')[-1]
            href = urlparse.urljoin('http://faculty.chicagobooth.edu/ruey.tsay/teaching/fts3/', a_ele.attrib['href']).strip()
            name = name.strip()
            
            file_path = os.path.join(_path, name)
            
            if href.find('/fts2/') > -1:
                if os.path.exists(file_path):
                    print 'remove: %s' % file_path
                    os.remove(file_path)
                print 'pass: %s: %s - %s' % (chapter, name, href)
                continue
            
            if not os.path.exists(file_path):
                print 'load: %s: %s - %s' % (chapter, name, href)
                
                try:
                    lr.load(href)
                    open(file_path, 'w').write(lr.body)
                except urllib2.HTTPError as e:
                    print e
                
            else:
                print 'pass: %s: %s - %s' % (chapter, name, href)
                
            
    
    
```
