```python
root = 'Y:/fin_data'
```

```python
for p in os.listdir(root):
    pp = os.path.join(root, p)
    if os.path.isdir(pp) and p >= '2023-06-12':
        for f in os.listdir(pp):
            ss = f.split('.')
            if len(ss) == 2:
                symbol = ss[0]
                exchange = None
                for _exchange, symbol_re in EXCHANGE_REGEX:
                    if symbol_re.search(symbol):
                        exchange = _exchange
                        break
                if exchange:
                    src_path = os.path.join(pp, f)
                    dst_path = os.path.join(pp, '%s.%s' % (exchange, f))
                    print('%s - %s' % (src_path, dst_path))
                    os.rename(src_path, dst_path)
            
        
        
```
