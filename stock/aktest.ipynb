```python
import akshare as ak
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import rcParams
```

```python
rcParams['figure.figsize'] = (20, 10)
```

```python
%matplotlib inline
```

```python
df = ak.stock_zh_a_hist(symbol='603636', period='daily', start_date='20100301', end_date='20220707', adjust='qfq')
```

```python
df['diff'] = df['收盘'].diff(-1).fillna(0)
```

```python
plt.scatter(x=df['diff'], y=range(df.shape[0]))
```

```python
f, ax = plt.subplots(figsize=(20, 10))
sns.histplot(df, x='diff', ax=ax)
```

```python
sns.set(rc={"figure.figsize":(20, 10)})
```

```python
plt.figure(figsize=(20,10))
sns.displot(df, x='diff', kind='kde')
# f, ax = plt.subplots(figsize=(20, 10))
# sns.kdeplot(df, x='diff', ax=ax)
```
